```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

def soccer_pitch(axs):
  rs={
    'x':[0,0,102,0,112],
    'y':[0,20,20,30,30],
    'w':[122,20,20,10,10],
    'h':[82,42,42,21,21]}
  plt.vlines(61,0,82,color='white')
  for n in range(5):
    axs.add_patch(Rectangle((rs['x'][n],rs['y'][n]),rs['w'][n],rs['h'][n],facecolor='Green',edgecolor='White'))
    w=Wedge((20,41),7.5,theta1=270,theta2=90,facecolor='Green', edgecolor='White')
    axs.add_patch(w)
    w=Wedge((102,41),7.5,theta1=90,theta2=270,facecolor='Green',edgecolor='White')
    axs.add_patch(w)
    c=Circle((61,41),10,facecolor='Green',edgecolor='White')
    axs.add_patch(c)
    axs.axis('off')
```

```python

fig, axs = plt.subplots(figsize=(6,4))
soccer_pitch(axs)
sns.scatterplot(data=data,x='start_location_x',y='start_location_y', ax=axs, hue='period')
plt.show()
```

```python
gb_df=data.groupby(['team','player','recipient']).count()['possession']
```

```python
gb_df=gb_df.reset_index()
gb_df
```

```python
pivot=pd.pivot(data=gb_df[gb_df.team=='France'],index='player',columns='recipient',values='possession')
pivot.fillna(0, inplace=True)
```

```python
sns.heatmap(pivot ,cmap='Reds',annot=True,fmt='G')
plt.title("Team plays for Frace")
plt.show()
```

```python
cols=['team','player','start_location_x', 'start_location_y']
gb_data2=data[cols].groupby(['team','player']).mean().reset_index()
gb_data2
```

```python
fig, axs = plt.subplots(figsize=(12,8))
soccer_pitch(axs)
sns.scatterplot(data=gb_data2,x='start_location_x',y='start_location_y',hue='team',size=gb_df['possession'])
for n in range(len(gb_data2)):
  axs.text(gb_data2['start_location_x'][n],gb_data2['start_location_y'][n],gb_data2['player'][n])
plt.legend(bbox_to_anchor=(1,1))
plt.show()

```
